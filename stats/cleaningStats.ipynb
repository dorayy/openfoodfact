```python
import streamlit as st
import pandas as pd
import requests
from pprint import pprint
import datetime
import numpy as np
```

```python
data = {'page_size': 50}
r1 = requests.get("https://world.openfoodfacts.org/products.json", params=data)
df = pd.DataFrame(r1.json()['products'])
# ,'nutriscore_data'
df = df[['id', 'product_name', 'nutriscore_grade' ,'expiration_date','scans_n' , 'stores']]
```

```python
# order by expiration date
df = df.sort_values(by=['expiration_date'], ascending=False)
# replace empty row at expiration date column by NaN
df['expiration_date'].replace('', np.nan, inplace=True)
df = df.dropna(subset=['expiration_date'])
df.head(50)
```
